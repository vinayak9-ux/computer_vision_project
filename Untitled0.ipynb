```python
df=pd.DataFrame(columns=['filename','object_num','bounding_box','shape'])
```

```python
def myfun(mystring):
    data=[]
    li=mystring.split()
    for i in range(0,len(li),5):
        data.append([li[i], [li[i+1],li[i+2],li[i+3],li[i+4]] ])
    return data
```

```python
for i in os.listdir(annotation_path):
    with open(annotation_path+"/"+i) as f:
        filname=i.split(".")[0]
        ext=''
        #print(filname)
        if os.path.isfile(img_path+"/"+filname+".png"):
            ext='.png'
        elif os.path.isfile(img_path+"/"+filname+".jpg"):
            ext='.jpg'
        img=cv.imread(img_path+"/"+filname+ext)
        data=myfun(f.read())
        for index,bounding_box in data:
            pass
            df.loc[len(df.index)] = [str(filname+ext), index, bounding_box,img.shape]
```

```python
df
```

```python
def yolo_to_voc_convertion(df):
    data=df['bounding_box']
    #print(data)
    image_height=df['shape'][0]
    image_width=df['shape'][1]

    x_center = float(data[0])
    y_center = float(data[1])
    width = float(data[2])
    height = float(data[3])

    x_min = int((x_center - (width / 2)) * image_width)
    y_min = int((y_center - (height / 2)) * image_height)
    x_max = int((x_center + (width / 2)) * image_width)
    y_max = int((y_center + (height / 2)) * image_height)

    return [x_min,y_min,x_max,y_max]
```

```python
df['new_bounding_box']=df.apply(yolo_to_voc_convertion,axis=1)
```

```python
#### filename- name of the file along with its extension
#### object_num- class index
#### bounding_box- normalized values of the bounding box generated by cvat
#### shape- image dimension
#### new_bounding_box-actual bounding box of the items in the image
```
